# Maplotlib implementation using rougha and real time dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4,5]
y=[0,2,4,6,8,10]
z = [3,6,9,12,15,18]
xticks = [1,2,3,4,5]
yticks = [2,4,6,8,16]
```

```python
plt.plot(x, y,'r.--',label = 'val', linewidth = 2,markersize = 10, markeredgecolor='violet')
plt.plot(x, z,label = 'val2',color='violet', linewidth = 2,marker = '.',markersize=10,markeredgecolor='purple',linestyle='--')

plt.title('Axis line', fontdict={ 'fontsize': 18})
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.xticks(np.arange(1,6,1))
plt.yticks(yticks)

plt.legend()
plt.show()
```

## instead of passing this all color , marker , line  mentioning it can be done in single set as [color][marker][line]--> 'r.--'



```python
x2 = np.arange(1,6,1)
plt.figure(figsize=(5,5),dpi=100)
plt.plot(x2, x2**2 , 'b^--',label = 'x2')
plt.plot(x2,x2*2,'r.--',label = 'x22')
plt.title('New Graph plot',fontdict={'fontsize':20})
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks([1,2,3,4,5])
plt.yticks([10,20,30])
plt.legend()
plt.savefig('Graph.png',dpi = 300)
plt.show()
```

#Bar Charts

```python
labels = ['A','B','C']
values = [1,4,2]
plt.figure(figsize=(6,4))

bars = plt.bar(labels,values)
plt.yticks([1,2,3,4])
plt.xticks(['A','B','C'],['Label A','Label B','Label C'])


bars[0].set_hatch('/')

bars[1].set_color('purple')

bars[1].set_hatch('/')
bars[2].set_hatch('*')

plt.show()
```

```python
labels = ['A','B','C','D']
values = [1,4,2,8]
patterns = ['/','*','o','.']
plt.figure(figsize=(6,4),dpi=150)

bars = plt.bar(labels,values,color='violet',label='bar chart')

for bar in bars:
  bar.set_hatch(patterns.pop(0))

plt.legend()

plt.show()

```

```python
x_labels = ['A', 'B', 'C']
y_values = [1, 2, 3]

plt.bar(x_labels, y_values)

# Customize y-axis tick labels
plt.yticks([1, 2, 3, 4], ['One', 'Two', 'Three', 'Four'])

# Customize x-axis tick labels
plt.xticks(x_labels, ['Label A', 'Label B', 'Label C'])

plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Customized Bar Chart')

plt.show()
```

#with real world examples

```python
gas = pd.read_csv('gas_prices.csv')
gas.sample(4)
```

```python


plt.figure(figsize = (8,5))

plt.plot(gas.Year,gas.USA,'r.-',label='USA',markeredgecolor='blue')
plt.plot(gas.Year,gas.Canada,'b.-',label='Canada',markeredgecolor='red')

plt.title('USA vs Canada Gas Prices')
plt.xlabel('Year')
plt.ylabel('Dollars/Gallon')
plt.legend()

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.show()

```

```python


plt.figure(figsize = (8,5))

for i in gas.columns[1:]:
 plt.plot(gas.Year,gas[i],label = i,marker='.')

plt.title('Gas prices',fontdict={'fontweight':'bold','fontsize':23})
plt.xlabel('Year')
plt.ylabel('Dollars/Gallon')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.show()

```

# Pie chart for percentage of people playing in left, right foot

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
```

```python
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
```

```python
labels = ['Left','Right']
colors = ['purple','violet']
plt.pie([left,right],labels = labels,colors=colors,autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light
```

```python
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
```

```python
light_medium
```

```python
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]
```

```python
medium,medium_heavy,heavy
```

```python
weights = (light,light_medium,medium,medium_heavy,heavy)
weights
```

```python

plt.style.use('ggplot')
explode = (.36,.27,0,0,.36)
plt.pie(weights,labels=('light','light_medium','medium','medium_heavy','heavy'),autopct='%.2f %%' ,pctdistance=0.9,explode=explode)
plt.title('Weight distribution of FIFA players (in lbs)')
plt.show()
```

#boxplot

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
```

```python
labels = ['FC Barcelona','Real Madrid']
color = ['purple','violet']

plt.style.use('ggplot')

boxes = plt.boxplot([barcelona,madrid],labels = labels,patch_artist=True,medianprops={'linewidth' : 2,'color' : 'red', 'linestyle' : '--'})

for box in boxes['boxes']:
    # Do something with each box
    box.set(color='purple',linewidth = 2,)
    box.set(facecolor = 'violet')

plt.title('Professional FIFA Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
